## Data Assessing

Mengecek kualitas data: missing values, duplicates, dan tipe data.

### Cek Missing Values

```python
print('=== Missing Values ===')
for name, df in [('customers', customers), ('orders', orders), ('order_items', order_items),
                  ('order_payments', order_payments), ('order_reviews', order_reviews),
                  ('products', products), ('sellers', sellers)]:
    nulls = df.isnull().sum()
    if nulls.sum() > 0:
        print(f'\n{name}:')
        print(nulls[nulls > 0])
```

### Cek Duplikat

```python
print('=== Duplikat ===')
for name, df in [('customers', customers), ('orders', orders), ('order_items', order_items),
                  ('order_payments', order_payments), ('order_reviews', order_reviews),
                  ('products', products), ('sellers', sellers)]:
    dups = df.duplicated().sum()
    if dups > 0:
        print(f'{name}: {dups} duplikat')
```

### Konversi Tipe Data

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
order_items['shipping_limit_date'] = pd.to_datetime(order_items['shipping_limit_date'])
order_reviews['review_creation_date'] = pd.to_datetime(order_reviews['review_creation_date'])
order_reviews['review_answer_timestamp'] = pd.to_datetime(order_reviews['review_answer_timestamp'])
print('Konversi datetime selesai')
```

## Data Cleaning

Membersihkan data: menangani missing values dan merge dataset.

### Merge Dataset Utama

Menggabungkan tabel orders dengan customers, order_items, payments, reviews, products, dan category_translation.

```python
# Merge orders with customers
main_df = orders.merge(customers, on='customer_id', how='left')

# Merge with order_items (aggregate per order)
items_agg = order_items.groupby('order_id').agg(
    total_items=('order_item_id', 'count'),
    total_price=('price', 'sum'),
    total_freight=('freight_value', 'sum'),
    product_ids=('product_id', lambda x: list(x))
).reset_index()
main_df = main_df.merge(items_agg, on='order_id', how='left')

# Merge with payments
payments_agg = order_payments.groupby('order_id').agg(
    payment_types=('payment_type', lambda x: list(x)),
    total_payment=('payment_value', 'sum')
).reset_index()
main_df = main_df.merge(payments_agg, on='order_id', how='left')

# Merge with reviews
reviews_agg = order_reviews.groupby('order_id').agg(
    review_score_mean=('review_score', 'mean'),
    review_count=('review_id', 'count')
).reset_index()
main_df = main_df.merge(reviews_agg, on='order_id', how='left')

# Merge with category translation
products_with_cat = products.merge(category_translation, on='product_category_name', how='left')
print(f'Main dataset shape: {main_df.shape}')
```

### Handle Missing Values

```python
print('Missing values:')
print(main_df.isnull().sum()[main_df.isnull().sum() > 0])
```

```python
# Fill missing values
main_df['review_score_mean'] = main_df['review_score_mean'].fillna(0)
main_df['review_count'] = main_df['review_count'].fillna(0)
main_df['total_payment'] = main_df['total_payment'].fillna(0)
main_df['total_items'] = main_df['total_items'].fillna(0).astype(int)
main_df['total_price'] = main_df['total_price'].fillna(0)
main_df['total_freight'] = main_df['total_freight'].fillna(0)
print('Missing values setelah cleaning:')
print(main_df.isnull().sum()[main_df.isnull().sum() > 0])
```

## Exploratory Data Analysis (EDA)

### Pertanyaan 1: Bagaimana distribusi status pesanan?

```python
order_status_counts = main_df['order_status'].value_counts()
print('Distribusi Order Status:')
print(order_status_counts)

plt.figure(figsize=(10, 6))
colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#6A994E', '#8D6B94', '#E9C46A', '#F4A261']
ax = sns.countplot(data=main_df, y='order_status', order=order_status_counts.index, palette=colors)
for i, v in enumerate(order_status_counts.values):
    ax.text(v + 200, i, str(v), va='center', fontsize=10)
plt.title('Distribusi Status Pesanan', fontsize=14, fontweight='bold')
plt.xlabel('Jumlah Pesanan')
plt.ylabel('Status')
plt.tight_layout()
plt.show()
```

### Pertanyaan 2: Bagaimana distribusi skor review?

```python
review_dist = order_reviews['review_score'].value_counts().sort_index()
print('Distribusi Review Score:')
print(review_dist)

plt.figure(figsize=(10, 6))
colors_review = ['#C73E1D' if s <= 2 else '#F18F01' if s == 3 else '#2E86AB' for s in review_dist.index]
ax = sns.barplot(x=review_dist.index, y=review_dist.values, palette=colors_review)
for i, v in enumerate(review_dist.values):
    ax.text(i, v + 500, f'{v}\n({v/review_dist.sum()*100:.1f}%)', ha='center', fontsize=10)
plt.title('Distribusi Skor Review', fontsize=14, fontweight='bold')
plt.xlabel('Skor Review')
plt.ylabel('Jumlah')
plt.tight_layout()
plt.show()
```

### Pertanyaan 3: Metode pembayaran apa yang paling populer?

```python
payment_type_counts = order_payments['payment_type'].value_counts()
print('Distribusi Metode Pembayaran:')
print(payment_type_counts)

plt.figure(figsize=(8, 6))
plt.pie(payment_type_counts.values, labels=payment_type_counts.index, autopct='%1.1f%%',
        colors=['#2E86AB', '#A23B72', '#F18F01', '#6A994E', '#C73E1D'], startangle=90)
plt.title('Distribusi Metode Pembayaran', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Pertanyaan 4: 10 kategori produk teratas berdasarkan penjualan?

```python
prod_sales = order_items.merge(products_with_cat, on='product_id', how='left')
cat_sales = prod_sales['product_category_name_english'].value_counts().head(10)
print('10 Kategori Produk Terlaris:')
print(cat_sales)

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=cat_sales.values, y=cat_sales.index, palette='viridis')
for i, v in enumerate(cat_sales.values):
    ax.text(v + 200, i, str(v), va='center', fontsize=9)
plt.title('10 Kategori Produk Terlaris', fontsize=14, fontweight='bold')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Kategori Produk')
plt.tight_layout()
plt.show()
```

### Pertanyaan 5: Bagaimana tren penjualan bulanan?

```python
main_df['purchase_month'] = main_df['order_purchase_timestamp'].dt.to_period('M')
monthly_sales = main_df.groupby('purchase_month').agg(
    order_count=('order_id', 'count'),
    total_revenue=('total_payment', 'sum')
).reset_index()
monthly_sales['purchase_month'] = monthly_sales['purchase_month'].astype(str)
print('Tren Penjualan Bulanan:')
print(monthly_sales.head(10))
```

```python
fig, ax1 = plt.subplots(figsize=(14, 6))
ax1.plot(monthly_sales['purchase_month'], monthly_sales['order_count'],
         marker='o', color='#2E86AB', linewidth=2, label='Jumlah Pesanan')
ax1.set_xlabel('Bulan')
ax1.set_ylabel('Jumlah Pesanan', color='#2E86AB')
ax1.tick_params(axis='x', rotation=45)
ax1.set_xticks(range(0, len(monthly_sales), 3))

ax2 = ax1.twinx()
ax2.plot(monthly_sales['purchase_month'], monthly_sales['total_revenue'] / 1e6,
         marker='s', color='#A23B72', linewidth=2, label='Revenue (Juta)')
ax2.set_ylabel('Revenue (Juta R$)', color='#A23B72')

plt.title('Tren Penjualan Bulanan', fontsize=14, fontweight='bold')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.95))
plt.tight_layout()
plt.show()
```

### Pertanyaan 6: Bagaimana distribusi waktu pengiriman?

```python
delivered = main_df[main_df['order_status'] == 'delivered'].copy()
delivered = delivered.dropna(subset=['order_delivered_customer_date'])
delivered['delivery_time_days'] = (
    delivered['order_delivered_customer_date'] - delivered['order_purchase_timestamp']
).dt.days
delivered['delivery_diff_estimated'] = (
    delivered['order_delivered_customer_date'] - delivered['order_estimated_delivery_date']
).dt.days
print('Statistik Waktu Pengiriman (hari):')
print(delivered['delivery_time_days'].describe())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(delivered['delivery_time_days'].clip(0, 50), bins=50,
             color='#2E86AB', edgecolor='white', alpha=0.8)
axes[0].axvline(delivered['delivery_time_days'].median(), color='#C73E1D',
                linestyle='--', linewidth=2,
                label=f'Median: {delivered["delivery_time_days"].median():.0f} hari')
axes[0].set_title('Distribusi Waktu Pengiriman')
axes[0].set_xlabel('Waktu Pengiriman (hari)')
axes[0].set_ylabel('Jumlah Pesanan')
axes[0].legend()

early = (delivered['delivery_diff_estimated'] < 0).sum()
late = (delivered['delivery_diff_estimated'] > 0).sum()
on_time = (delivered['delivery_diff_estimated'] == 0).sum()
axes[1].pie([early, on_time, late], labels=['Lebih Cepat', 'Tepat Waktu', 'Terlambat'],
            autopct='%1.1f%%', colors=['#2E86AB', '#6A994E', '#C73E1D'], startangle=90)
axes[1].set_title('Ketepatan Waktu Pengiriman')

plt.tight_layout()
plt.show()
```

### Pertanyaan 8: Hubungan harga dan skor review?

```python
price_review = order_items.groupby('order_id')['price'].sum().reset_index()
price_review = price_review.merge(
    order_reviews.groupby('order_id')['review_score'].mean().reset_index(),
    on='order_id', how='inner'
)
print('Korelasi Harga vs Review Score:')
print(price_review[['price', 'review_score']].corr())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=price_review.sample(5000), x='price', y='review_score',
                alpha=0.3, color='#2E86AB')
plt.title('Hubungan Harga vs Skor Review')
plt.xlabel('Total Harga Pesanan (R$)')
plt.ylabel('Rata-rata Skor Review')
plt.tight_layout()
plt.show()
```
